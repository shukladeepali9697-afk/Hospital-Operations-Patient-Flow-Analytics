```python
hospital_data['Admission_Date'] = pd.to_datetime(hospital_data['Admission_Date'])
```

```python
hospital_data['Discharge_Date'] = pd.to_datetime(hospital_data['Discharge_Date'])
```

```python
hospital_data.info()
```

```python
hospital_data['Admission_Date'].isnull().sum()
```

```python
hospital_data["BMI_Category"] = pd.cut(hospital_data["BMI"],bins=[0,18.5,25,30,100], labels=["Underweight","Normal","Overweight","Obese"])
```

```python
hospital_data['Age_Group'] = pd.cut(hospital_data['Age'], bins=[0,18,35,50,65,100], labels=["0-18","19-35","36-50","51-65","65+"])
```

```python
hospital_data['Billing_Category'] = pd.cut(hospital_data['Billing_Amount'], bins=[0,10000,25000,50000,100000], labels=["Low","Medium","High","Very High"])
```

```python
hospital_data['Stay_Category'] = pd.cut(hospital_data['Days_Hospitalised'], bins=[0,5,10,20,50], labels=["Short Stay","Medium Stay","Long Stay","Critical Stay"])
```

```python
hospital_data['Discharge_Year'] = (hospital_data['Discharge_Date'].dt.year)
```

```python
hospital_data['Discharge_Month'] = (hospital_data['Discharge_Date'].dt.month_name())
```
